### 4.3. Shapiro-Wilk Test

The Shapiro-Wilk test is used for testing the normality of a dataset. The null hypothesis in this test is that the population is normally distributed. Therefore, when performed, if the p-value is less than the chosen significance (alpha), the null-hypothesis is rejected and there is evidence that the data tested is not normally distributed. On the other hand, if the p-value is greater than the chosen significance, the null-hypothesis cannot be rejected and there is evidence that the data tested is normally distributed.

Assessing the normality of a dataset is important because it allows us to understand underlying premises on the data and choose better strategies accordingly.

```python
from scipy.stats import shapiro

alpha = 0.05

shapiro_results = {'checkout_1': {'statistic': shapiro(checkout_1)[0], 'p-value': shapiro(checkout_1)[1]},
                   'checkout_2': {'statistic': shapiro(checkout_2)[0], 'p-value': shapiro(checkout_2)[1]}}

print(f"Shapiro-Wilk Test - checkout_1: Statistic = {shapiro_results['checkout_1']['statistic']:.5f}")
print(f"Shapiro-Wilk Test - checkout_2: Statistic = {shapiro_results['checkout_2']['statistic']:.5f}")
print("--------------")

print(f"p-value - checkout_1 = {shapiro_results['checkout_1']['p-value']:.5f}")
print(f"p-value - checkout_2 = {shapiro_results['checkout_2']['p-value']:.5f}")
print("--------------")

condition_1 = shapiro_results['checkout_1']['p-value'] > alpha
condition_2 = shapiro_results['checkout_2']['p-value'] > alpha

if condition_1:
    if condition_2:
        print('Both datasets appear to be normally distributed.')
    else:
        print('checkout_1 appears to be normally distributed. However, checkout_2 does not appear to be normally distributed.')
else:
    if condition_2:
        print('checkout_2 does not appear to be normally distributed. However, checkout_2 appears to be normally distributed.')
    else:
        print('Neither dataset appears to be normally distributed. Null hypothesis rejected.')
```

## 5. Outlier detection

To detect outliers in "today" data, a custom anomaly score was designed, based on eight different tests. First, it was decided that the main information needed for the tests is the difference between today's value and each of the other columns corresponding value for each different hour of the dataset. With this difference, we can examine the patterns and see if there are any significant deviations between today's behavior and the other days and average values.

Since there are 4 columns besides "today" (which are "yesterday", "same_day_last_week", "avg_last_week", and "avg_last_month"), 4 different variables will come up, referring to the respective deviation between today and each of the variables. Additionally, 2 different outlier detection techniques will be applied to each of the 4 new variables. The first technique consists is statistical and is based on the standard deviation value for the original column that we are comparing today's results to. For example, if we want to detect outliers using data from the average values for last month, we would inspect the values for "today - avg_last_month" and see which of them represent a deviation that is higher than the standard deviation for the average of last month multiplied by a value, i.e., 1.5.

The second test is a Machine Learning technique known as Isolation Forest, and it uses an ensemble of decision trees to split up the data and isolate each data point. The data points that are easily (with few steps) isolated from the other ones might be considered outliers, and the sensitivity of the detection depends on a "contamination" parameter.

Since 2 different tests will be applied to each of the 4 new variables, this leads up to a total of 8 different tests per data point (hour), and the final anomaly score will be computed as the number of tests among the 8 in which that point was detect as an outlier. Therefore, the anomaly scores will range from 0 to 8.

```python
def compute_anomaly_scores(df, threshold = 1.5, columns = ('avg_week','avg_month','yesterday','same_day_last_week')):

    df.columns = ['today', 'yesterday', 'same_day_last_week', 'avg_week',
       'avg_month']
    
    for col in columns:
        df[f'today_{col}_diff'] = df['today'] - df[col]
        
    
    means = {col: df[f'today_{col}_diff'].mean() for col in columns}
    stds = {col: df[f'today_{col}_diff'].std() for col in columns}
    anomalies = {col: (df[f'today_{col}_diff'].abs() > threshold*stds[col]) for col in columns}
    colors = {col: np.where(anomalies[col], 'r', 'g') for col in columns}
    
    models = {}
    IF_anomalies = {}

    for col in columns:
        data = df[[f'today_{col}_diff']]
        model = IsolationForest(n_estimators = 100, contamination='auto')
        model.fit(data)
        prediction = model.predict(data)
        anomaly = prediction == -1
        models[col] = model
        IF_anomalies[col] = anomaly
        
    IF_dataframe = pd.DataFrame(IF_anomalies)
    IF_dataframe[IF_dataframe == True] = 1
    IF_dataframe[IF_dataframe == False] = 0

    dataframe = pd.DataFrame(anomalies)
    dataframe[dataframe == True] = 1
    dataframe[dataframe == False] = 0
    IF_dataframe.index = dataframe.index

    dataframe['threshold_score'] = dataframe.sum(axis=1)
    IF_dataframe['IF_score'] = IF_dataframe.sum(axis=1)
    total_score = pd.concat([dataframe['threshold_score'],IF_dataframe['IF_score']], axis=1)
    
    score_total = total_score.sum(axis=1)
    
    return score_total, anomalies, IF_anomalies

def plot_anomalies(df, score_total, data = 'checkout_1', columns = ('avg_week','avg_month','yesterday','same_day_last_week')):
                   
    fig, ax = plt.subplots(figsize=(10,3))
                   
    colors = np.where(score_total > 5, 'r',
                  np.where(score_total > 3, 'orange',
                          np.where(score_total > 2, 'yellow', 'green')))
                   
    
    ax.bar(score_total.index, score_total, color = colors, alpha = 0.5)
    ax.set_xlim(-0.5, 23.5)    
    ax.set_ylabel('Anomaly score')
    ax.set_title(f'Anomaly detection for {data}')
    
    ax2 = ax.twinx()
    ax2.plot(checkout_1.index, checkout_1['today'], c = 'k', lw = 1.0, label = 'Today')
    ax2.set_ylabel('Today values')    
    

    
    # Add legend for bar plot
    legend_labels = ['Score  6', '4  Score < 6', 'Score = 3', 'Score  2']
    legend_colors = ['r', 'orange', 'yellow', 'green']
    legend_handles = [plt.Rectangle((0,0),1,1, color=color, label=label) for color, label in zip(legend_colors, legend_labels)]
    
    bar_legend = ax.legend(handles=legend_handles, loc='upper left')
    line_legend = ax2.legend(loc='upper right')
    
    ax.add_artist(bar_legend)
    ax2.add_artist(line_legend)
    
    
def general_plot_with_anomalies(df, score_total, data):
    
    from matplotlib.lines import Line2D
    
    columns = ['today','avg_week','avg_month','yesterday','same_day_last_week']
    
    colors = np.where(score_total > 5, 'r',
                  np.where(score_total > 3, 'orange',
                          np.where(score_total > 2, 'yellow', 'green')))   
    sizes = np.where(score_total > 5, 120,
                  np.where(score_total > 3, 80,
                          np.where(score_total > 2, 50, 25)))
    
    fig, ax = plt.subplots(figsize=(10,5))
    
    line_colors = ['black', 'blue','green','red','purple']
    line_styles = ['-','-','--','-.',':']
    
    ## Plot scatter plot
    scatter_handles = []  # List to store legend handles for scatter markers
    for color, label in zip(['r', 'orange', 'yellow', 'green'], ['Score  6', '4  Score < 6', 'Score = 3', 'Score  2']):
        scatter = Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label)
        scatter_handles.append(scatter)
    
    for i, col in enumerate(columns):
        ax.plot(df.index, df[col], lw = 1.5, label = col, color= line_colors[i], ls = line_styles[i])

    ax.scatter(df.index, df['today'], c = colors, s = sizes, zorder = 10)
    ax.set_title(f'Comparative plot with anomaly detections for {data}') 
    
    # Add legend for scatter plot
    scatter_legend = ax.legend(handles=scatter_handles, loc='upper left')

    # Add legend for line plots
    line_legend = ax.legend(loc='upper right')
    
    # Add both legends to the plot
    ax.add_artist(scatter_legend)
    ax.add_artist(line_legend)
    
```

```python
score_total1, anomalies1, IF_anomalies1 = compute_anomaly_scores(checkout_1, threshold = 1.5)
score_total2, anomalies2, IF_anomalies2 = compute_anomaly_scores(checkout_2, threshold = 1.5)
```

```python
plot_anomalies(checkout_1, score_total1, data = 'checkout_1')
plot_anomalies(checkout_2, score_total2, data = 'checkout_2')
```

```python
general_plot_with_anomalies(checkout_1, score_total1, 'checkout_1')
general_plot_with_anomalies(checkout_2, score_total2, 'checkout_2')
```

## 6. Assigning anomalies and levels

```python
for df, score_total in zip([checkout_1, checkout_2],[score_total1, score_total2]):
    
    df['Anomaly score'] = score_total    
    df['Anomaly level'] = np.where(score_total > 5, 'Severe',
                            np.where(score_total > 3, 'High',
                                    np.where(score_total > 2, 'Possible', 'Not an anomaly')))
    df['Anomaly detection'] = np.where(score_total > 3, True, False)
```
