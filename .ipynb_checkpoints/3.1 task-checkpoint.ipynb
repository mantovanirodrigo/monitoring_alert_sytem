# Statistical Thresholding

```python
df = checkout_1

row_means = df.mean(axis=1)
row_medians = df.median(axis=1)
row_std = df.std(axis=1)
row_mad = df.mad(axis=1)

threshold = 2
anomalies_mean = (df['today'] - row_means).abs() > threshold * row_std
anomalies_median = (df['today'] - row_medians).abs() > threshold * row_mad
labels = ['Today','Yesterday','Same day (last week)','Avg. (last week)','Avg. (last month)']

fig, ax = plt.subplots(2, 1, figsize=(10,7))
plt.subplot(2, 1, 1)

ax[0].plot(df.index, df, label = labels)
anomaly_indices = np.where(anomalies_median)[0]  # Get indices of anomalies
ax[0].plot(df.index[anomaly_indices], df['today'].iloc[anomaly_indices], 'ro', label='Anomalies')  # Plot anomalies with red dots
ax[0].legend()
ax[0].set_title(f'Median anomaly detection')

ax[1].plot(df.index, df, label = labels)
anomaly_indices = np.where(anomalies_mean)[0]  # Get indices of anomalies
ax[1].plot(df.index[anomaly_indices], df['today'].iloc[anomaly_indices], 'ro', label='Anomalies')  # Plot anomalies with red dots
ax[1].legend()
ax[1].set_title(f'Mean anomaly detection')
plt.tight_layout()
```

```python
anomalies_both = np.where(anomalies_mean & anomalies_median)[0]
checkout_1.iloc[np.where(anomalies_median)[0]]
```

```python
from scipy.stats import shapiro
alpha = 0.05

shapiro_dict = {}
for i, col in enumerate(diff_df.columns):
    stat, p = shapiro(diff_df[col])
    shapiro_dict[col] = {'statistic': stat,
                         'p-value': p,
                        'is normal': True}
for col in diff_df.columns:
    if shapiro_dict[col]['p-value'] < alpha:
        shapiro_dict[col]['is normal'] = False    

pd.DataFrame(shapiro_dict).transpose()
```

```python
from scipy.stats import kstest, norm
alpha = 0.05

diff_df_norm = (diff_df - diff_df.mean())/diff_df.std() 


kstest_dict = {}
for i, col in enumerate(diff_df.columns):
    stat, p = kstest(diff_df_norm[col], 'norm')
    kstest_dict[col] = {'statistic': stat,
                         'p-value': p,
                        'is normal': True}
for col in diff_df.columns:
    if kstest_dict[col]['p-value'] < alpha:
        kstest_dict[col]['is normal'] = False    

pd.DataFrame(kstest_dict).transpose()
```

```python
from scipy.stats import shapiro

alpha = 0.05

shapiro_result_today_avgmonth = shapiro(diff_df['today - avg_last_month'])
shapiro_result_today_avgweek = shapiro(diff_df['today - avg_last_week'])

print(f"Shapiro-Wilk Test - Pivs Centrais: Statistic={shapiro_result_today_avgmonth[0]}")
print(f"Shapiro-Wilk Test - Colheitadeiras: Statistic={shapiro_result_today_avgweek[0]}")

print(f"P-valor (pivs) = {shapiro_result_today_avgmonth[1]}")
print(f"P-valor (colheitadeiras) = {shapiro_result_today_avgweek[1]}")
print("--------------")

if shapiro_result_today_avgmonth[1] < alpha and shapiro_result_today_avgweek[1] < alpha:
    
    print("Data does not appear to be normally distributed. Consider using Spearman's correlation")

else: 
    
    print("Data appears to be normally distributed. Pearson's correlation can be used")
```

```python
shapiro_test = stats.shapiro(diff_df['today - avg_last_month'])
shapiro_test_c = stats.shapiro(diff_df['today - avg_last_week'])
print(f"Pivs Centrais - Shapiro-Wilk Test: W={shapiro_test[0]:.4f}, p-value={shapiro_test[1]:.4f}")
print(f"Colheitadeiras - Shapiro-Wilk Test: W={shapiro_test_c[0]:.4f}, p-value={shapiro_test_c[1]:.4f}")
```
